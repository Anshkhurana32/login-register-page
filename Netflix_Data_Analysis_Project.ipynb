Dropping the Overview,Original_Language,	Poster_Url

```python
col=['Overview','Original_Language', 'Poster_Url']
pf.drop(col,axis=1,inplace=True)
```

```python
pf
```

```python
def categorize_col(pf, col, labels):
    edge = [
        pf[col].describe()['min'],
        pf[col].describe()['25%'],
        pf[col].describe()['50%'],
        pf[col].describe()['75%'],
        pf[col].describe()['max']
    ]

    pf[col] = pd.cut(pf[col], bins=edge, labels=labels, duplicates='drop')
    return pf


```

```python
label=['not_popular','below_avg','average','popular']
categorize_col(pf,'Vote_Average',label)
```
